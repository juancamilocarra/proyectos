```python
y= df["viaje_noche_fuera"]
x= df[["edad","estado_civil"]]
```

```python
y=y.replace({"si":1,"no":0})
```

```python
x=pd.get_dummies(x,drop_first=True)
```

```python
x = sm.add_constant(x)
model = sm.OLS(y, x)
reg = model.fit()
reg.summary()
```
